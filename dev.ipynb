```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
df = pd.read_csv('epa-sea-level.csv')
df.head()
```

```python
# draw the scatter plot
fig, ax = plt.subplots(figsize=(10,5))
ax.scatter(df['Year'], df['CSIRO Adjusted Sea Level'])
```

```python
lr = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
```

```python
x = pd.concat([df['Year'], pd.Series(2050)], ignore_index=True)
```

```python
ax.plot(x, lr.slope * x + lr.intercept)
fig
```

```python
# create a copy of the datframe with just the data from 2000 on
df2 = df[df['Year'] >= 2000]
```

```python
lr2 = linregress(df2['Year'], df2['CSIRO Adjusted Sea Level'])
lr2
```

```python
x2 = pd.concat([df2['Year'], pd.Series(2050)], ignore_index=True)
```

```python
ax.plot(x2, lr2.slope * x2 + lr2.intercept)
fig
```

```python
ax.set_xlabel('Year')
ax.set_ylabel('Sea Level (inches)')
ax.set_title('Rise in Sea Level')
fig
```
